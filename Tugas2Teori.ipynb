```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter simulasi
tahun_awal = 2023
tahun_akhir = 2030
jumlah_sekolah_awal = 50  # Misal jumlah sekolah dengan laboratorium pada tahun 2023
rata2_pertumbuhan = 0.05  # 5% pertumbuhan rata-rata
deviasi_standar = 0.02    # 2% deviasi standar
simulasi = 10000          # Jumlah simulasi

# Menyimpan hasil simulasi
hasil_simulasi = np.zeros((simulasi, tahun_akhir - tahun_awal + 1))

# Melakukan simulasi
for i in range(simulasi):
    jumlah_sekolah = jumlah_sekolah_awal
    for tahun in range(tahun_akhir - tahun_awal + 1):
        pertumbuhan = np.random.normal(rata2_pertumbuhan, deviasi_standar)
        jumlah_sekolah += jumlah_sekolah * pertumbuhan
        hasil_simulasi[i, tahun] = jumlah_sekolah

# Menghitung rata-rata dan interval kepercayaan
rata2_tiap_tahun = np.mean(hasil_simulasi, axis=0)
perc_5 = np.percentile(hasil_simulasi, 5, axis=0)
perc_95 = np.percentile(hasil_simulasi, 95, axis=0)

# Plot hasil
tahun = np.arange(tahun_awal, tahun_akhir + 1)
plt.fill_between(tahun, perc_5, perc_95, color='b', alpha=0.2, label='5th-95th Percentile')
plt.plot(tahun, rata2_tiap_tahun, 'b-', label='Rata-rata Prediksi')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Sekolah dengan Laboratorium')
plt.title('Prediksi Jumlah SD dan SMP dengan Laboratorium di Kota Cirebon')
plt.legend()
plt.grid(True)
plt.show()
```
